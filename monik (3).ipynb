```python
from scipy import stats
import numpy as np
```

```python
z_scores=np.abs(stats.zscore(df['Magnesium']))
df[z_scores>3]
```

```python
Q1=df['Magnesium'].quantile(0.25)
Q3=df['Magnesium'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
df[(df['Magnesium']<lower)|(df['Magnesium']>lower)]
```

```python
df['Magnesium']=np.where(df['Magnesium']>upper,upper,df['Magnesium'])
df['Magnesium']=np.where(df['Magnesium']<lower,lower,df['Magnesium'])
```

```python
df['Magnesium']=np.log(df['Magnesium'])
```
